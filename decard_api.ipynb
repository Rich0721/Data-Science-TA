```python
import requests, bs4, json
import pandas as pd
url_popular = 'https://www.dcard.tw/_api/posts?popular=true'
posts = list(requests.get(url_popular).json())
counter = 0

def printing():                         
    global counter
    df = {"標題":[], "內文":[]}
    for post in posts:
        counter += 1
        #print('貼文熱門編號 : ', counter) 
        df["標題"].append(post['title'])
        df["內文"].append(post['excerpt'])
    return df
       
df = printing()
print(df)
df = pd.DataFrame(df)
df.to_excel("decard.xlsx", index=False)
```
